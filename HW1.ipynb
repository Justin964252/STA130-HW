#Q6

definitions: Count: The number of non-missing (non-NaN) values in each column. It shows how many valid (non-null) data points are present for each variable.
Mean: The average value of the data in each column, calculated by summing all the valid values and dividing by the count. It represents the central tendency or "typical" value.
Std (Standard Deviation): A measure of the spread or dispersion of the data. It indicates how much the values deviate, on average, from the mean. A higher standard deviation means the data points are more spread out from the mean.
Min (Minimum): The smallest value in the dataset for each variable. This indicates the lower boundary of the dataset.
25% (First Quartile or Q1): The value below which 25% of the data falls. It represents the first quartile, giving you a sense of the lower part of the data distribution.
50% (Median or Second Quartile): The middle value of the dataset when sorted. Half of the data falls below this value, and half is above it. The median is less sensitive to extreme values than the mean.
75% (Third Quartile or Q3): The value below which 75% of the data falls. It gives insight into the upper part of the data distribution.
Max (Maximum): The largest value in the dataset for each variable. This indicates the upper boundary of the dataset.

These summary statistics (count, mean, std, min, 25%, 50%, 75%, and max) can only be calculated for numeric variables because they involve mathematical operations that require numerical values.
